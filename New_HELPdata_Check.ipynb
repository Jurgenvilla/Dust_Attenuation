```python
test = Table.read('/home/jvilla/Documents/PhD_2018/3DHST_Data/'
                  'COSMOS/HELP_Veronique_data/Help_cosmos_sample_herschel-snr5.fits', format = 'fits')
```

```python
test = test.to_pandas()
test = test.fillna(-9999.99)
test = Table.from_pandas(test)
```

```python
print('Band: \t\t\t Data in Vero: \t\tNon_Zero:\tMy data:\tNon_zero:')
for i in range(len(HELP_bands)):
    
    x = test[Buat_bands[i]][test[Buat_bands[i]] != -9999.99]
    y = test[HELP_bands[i]][test[HELP_bands[i]] != -9999.99]
    
    print('%20s\t\t%i\t\t%i\t\t%i\t\t%i'\
          %(Buat_bands[i], len(test[Buat_bands[i]]), len(x),\
            len(test[HELP_bands[i]]), len(y)))
```

```python
band_vero  = ['spitzer.mips.24',\
               'herschel.pacs.100', 'herschel.pacs.160',\
               'herschel.spire.PSW', 'herschel.spire.PMW', 'herschel.spire.PLW']

band_err_vero = ['spitzer.mips.24_err',\
                 'herschel.pacs.100_err', 'herschel.pacs.160_err',\
                 'herschel.spire.PSW_err', 'herschel.spire.PMW_err', 'herschel.spire.PLW_err']

band_mine = ['help_f_mips_24',\
             'help_f_pacs_green','help_f_pacs_red',\
             'help_f_spire_250', 'help_f_spire_350','help_f_spire_500']

band_err_mine = ['help_f_mips_24',\
                 'help_f_pacs_green','help_f_pacs_red',\
                 'help_f_spire_250', 'help_f_spire_350','help_f_spire_500']
```
